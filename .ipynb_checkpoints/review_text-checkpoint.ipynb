```python
import pandas as pd
import os
import nltk
```

```python
data_folder = 'Data/'
```

```python
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]
```

```python
movie_data = {}

for file_name in csv_files:
    movie_name = os.path.splitext(file_name)[0]
    df = pd.read_csv(os.path.join(data_folder, file_name))
    movie_data[movie_name] = df    
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    words = word_tokenize(text.lower())
    cleaned_words = [word for word in words if word.isalnum() and word not in stop_words]
    return ' '.join(cleaned_words)
```

```python
for movie_name, df in movie_data.items():
    df['review'] = df['review'].apply(preprocess_text)
```

```python
movie_corpus = {}

for movie_name, df in movie_data.items():
    movie_corpus[movie_name] = ' '.join(df['review'])
```
