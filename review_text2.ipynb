```python
import pandas as pd
import os
import nltk
impor
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
data_folder = 'Data/'
```

```python
# List all CSV files in the folder
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]
```

```python
# Create a dictionary to store DataFrames
movie_data = {}
```

```python
for file_name in csv_files:
    movie_name = os.path.splitext(file_name)[0]  # Extract movie name from filename
    df = pd.read_csv(os.path.join(data_folder, file_name))
    movie_data[movie_name] = df
```

```python
# Assuming you have NLTK installed
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    words = word_tokenize(text.lower())
    cleaned_words = [word for word in words if word.isalnum() and word not in stop_words]
    return ' '.join(cleaned_words)
```

```python
# Apply preprocessing to each review
for movie_name, df in movie_data.items():
    df['review'] = df['review'].apply(preprocess_text)
```

```python
movie_corpus = {}

for movie_name, df in movie_data.items():
    movie_corpus[movie_name] = ' '.join(df['review'])
```

```python
# Assuming you have a movie name, for example 'Toy Story 1995'
movie_name = 'Toy Story 1995'

# Print the corpus for the specified movie
print(movie_corpus[movie_name])
```

```python
# Convert movie_corpus to a DataFrame
df_movie_corpus = pd.DataFrame(list(movie_corpus.items()), columns=['Movie Name', 'Corpus'])

# Save the DataFrame to a CSV file
df_movie_corpus.to_csv('movie_corpora.csv', index=False)
```
